```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.info()
```

```python
df.drop(['Cabin'],axis= 1,inplace = True)
```

```python
df.fillna({'Age':df['Age'].mean()},inplace = True)
df.info()
```

```python
df.fillna(method = 'ffill',inplace = True)
df.info()
```

```python
df.drop(['PassengerId','Ticket'],axis = 1,inplace=True)
df.head()
```

```python
count = df['Embarked'].value_counts()
print(count)
```

```python
df['Sex'] = df['Sex'].replace(['male','female'],[1,0])
df.head()
```

```python
df['Embarked'] = df['Embarked'].replace(['Q','C'],['Other','Other'])
df.head()
```

```python
counts = df['Embarked'].value_counts()
print(counts)
```

```python
df = pd.get_dummies(df,columns = ['Embarked'],prefix='E')
df.head()
```

```python
counts = df['Parch'].value_counts()
print(counts)
```

```python
df['Parch'] = df['Parch'].replace([5,6,3,4,2],[1,1,1,1,1])
df.head()
```

```python
df = pd.get_dummies(df,columns = ['Parch'],prefix='Parch')
df.head()
```

```python
df['Title']=df.Name.apply(lambda name:name.split(',')[1].split('.')[0].strip() )
```

```python
## title counts
#print("There are {} unique title.".format(titanic.Title.nunique))
print("There are {} unique titles.".format(df.Title.nunique()))
print("\n", df.Title.unique())
```

```python
# normalize the titles
normalized_titles = {
    "Capt":       "Officer",
    "Col":        "Officer",
    "Major":      "Officer",
    "Jonkheer":   "Royalty",
    "Don":        "Royalty",
    "Sir" :       "Royalty",
    "Dr":         "Officer",
    "Rev":        "Officer",
    "the Countess":"Royalty",
    "Dona":       "Royalty",
    "Mme":        "Mrs",
    "Mlle":       "Miss",
    "Ms":         "Mrs",
    "Mr" :        "Mr",
    "Mrs" :       "Mrs",
    "Miss" :      "Miss",
    "Master" :    "Master",
    "Lady" :      "Royalty"
}

def convert(val):
    return normalized_titles[val]
```

```python
df.Title = df.Title.map(normalized_titles)
```

```python
# view value counts for the normalized titles
print(df.Title.value_counts())
```

```python
df.head()
```

```python
df.drop(['Name'],axis= 1,inplace = True)
```

```python
df = pd.get_dummies(df,columns = ['Title'],prefix='Title')
df.head()
```

```python
df = df.replace([True,False],[1,0])
df.head()
```

```python
x = df.drop(['Survived'],axis=1)
y= df['Survived']
x.head()
```

```python
tx,tex,ty,tey= train_test_split(x,y,random_state = 0,test_size = 1 )
```

```python
LogR_model=LogisticRegression()
```

```python
LogR_model.fit(tx,ty)
```

```python
LogR_predict =LogR_model.predict(tex)
```

```python
accuracy_score(LogR_predict,tey)
```

```python
def titanic(df):
    df.drop(['Cabin','PassengerId','Ticket'],axis= 1,inplace = True)
    df['Title']=df.Name.apply(lambda name:name.split(',')[1].split('.')[0].strip() )
    df.Title = df.Title.map(normalized_titles)
    df.drop(['Name'],axis= 1,inplace = True)
    
    df.fillna({'Age':df['Age'].mean()},inplace = True)
    df.fillna(method = 'ffill',inplace = True)
    df['Sex'] = df['Sex'].replace(['male','female'],[1,0])
    df['Embarked'] = df['Embarked'].replace(['Q','C'],['Other','Other'])
    df = pd.get_dummies(df,columns = ['Embarked'],prefix='E')
    df['Parch'] = df['Parch'].replace([5,6,3,4,2,9],[1,1,1,1,1,1])
    df = pd.get_dummies(df,columns = ['Parch'],prefix='Parch')
    df = pd.get_dummies(df,columns = ['Title'],prefix='Title')
    df = df.replace([True,False],[1,0])
    return df
test = titanic(test)
test.head()
```
